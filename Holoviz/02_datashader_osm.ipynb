```python
import pandas as pd
import datashader as ds
from datashader.utils import export_image
```

#### Constants

```python
background = "#242124"
cmap = ["indigo", "yellow", "white"]
```

#### Read the raw data

```python
df = pd.read_parquet("../Data/osm_gps_rome_raw.parquet")
```

#### Plot all the datapoints

```python
cvs = ds.Canvas(plot_width=1200, plot_height=1200)
agg = cvs.points(df, 'lon', 'lat')
img = ds.tf.shade(agg, cmap=cmap, how="log")
ds.tf.set_background(img, background)

# Remove the image border
ds.transfer_functions.Image.border = 0
```

#### Save the image as PNG

```python
export_image(img, "bg", background=background, export_path=".", fmt=".png")
```
